```python
from nltk.translate.bleu_score import sentence_bleu

reference = [['this', 'is', 'a', 'test']]
candidate = ['this', 'is', 'test']
score = round(sentence_bleu(reference, candidate, weights=(1, 1, 0, 0)),2)

print("BLEU Score:", score)
```

```python
from rouge import Rouge
rouge = Rouge()
scores = rouge.get_scores("The cat sat on the mat.", "The cat is on the mat.")
for score in scores:
    print("ROUGE-1:", score['rouge-1'])
    print("ROUGE-2:", score['rouge-2'])
    print("ROUGE-L:", score['rouge-l'])
```

```python
from rouge import Rouge

rouge = Rouge()
scores = rouge.get_scores("The cat sat on the mat.", "The cat is on the mat.")

for score in scores:
    for metric in ['rouge-1', 'rouge-2', 'rouge-l']:
        rounded = {k: round(v, 2) for k, v in score[metric].items()}
        print(f"{metric.upper()}:", rounded)
```
