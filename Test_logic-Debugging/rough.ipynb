```python
from langchain_core.messages import HumanMessage, AIMessage

# Simulate your state dict
state = {
    "messages": [
        HumanMessage(content="who wrote the mahabharat granth?"),
        AIMessage(content="The Mahabharata is traditionally attributed to Vyasa.")
    ]
}

# Debugging: print the whole state
print("Full state:", state)

# Get the last message object
last_message = state["messages"][-1]
print("Last message object:", last_message)

# Get just the text content of the last message
print("Last message content:", last_message.content)

# If you want to see all contents in order
for i, msg in enumerate(state["messages"]):
    print(f"Message {i}: role={msg.type}, content={msg.content}")
```
